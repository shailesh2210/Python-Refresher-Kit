```python
with open("data.txt", "r") as f:
    content = f.read()
    print(content)
```

```python
# reading line by line

with open("data.txt", "r") as f:
    for line in f:
        print(line.strip())
```

```python
# Read the first N characters

with open("data.txt", "r") as f:
    print(f.read(10))
```

```python
with open("data.txt", "w") as f:
    f.write("Hello Python\n")
    f.write("Revisting the file handling\n")
```

```python
# Appending (mostly used for logs)

with open("data.txt", "a") as f:
    f.write("New Entry Added\n")
```

```python
# Checking if file exists

import os

if os.path.exists("data.txt"):
    print("File Exists")
else:
    print("No file Found")
```
