```python
# conditional statements with input


age = int(input("Enter your age: "))

if age < 0:
    print("Invalid age")

elif age < 18:
    print("You're an minor")

elif age<=18:
    print("You're an adult")
else:
    print("You are a senior citizen")

```

```python
# Logical + Conditional operators (and, or, not)

age = int(input("Enter your age: "))

has_id = input("Do you have valid ID? (yes/no)").strip().lower()

if age >=18 and has_id == "yes":
    print("Entry Allowed ")

else:
    print("Entry not allowed")
```

```python
# Project: Simple Customer Segmentation using Conditional Statements

# we use everything that we have seen above and create an mini projecct using that 

"""We use:
- input()
- int() and str handling
- comparison operators
- logical operators (and, or)
- nested if
- one-line conditional (ternary)

This feels like a tiny rule-based model.
"""

print("==== Customer Segmentation Project ====")

name = input("Enter your name: ").strip()
age = int(input("Enter your age: "))
monthly_spend = int(input("Enter your monthly spend : "))
is_member = input("is the customer a member>=? (yes/no): ").strip().lower()
city = input("Enter the City: ").strip().lower()

# Determine value sigment 

if monthly_spend >=50000 and is_member == "yes":
    segment = "High Value"
elif monthly_spend >=20000:
    segment = "Medium Value"
else:
    segment = "Low Value"

#2 determine type by age

if age < 22:
    customer_type = "Student"
elif age <=25:
    customer_type = "Working Professional"
else:
    customer_type = "Senior"

priority_city = "Yes" if city in ["mumbai", "bangalore", "delhi", "hyderabad"] else "No"

print("\n=== Customer Summary ===")
print(f"Name          : {name}")
print(f"Age           : {age}")
print(f"City          : {city.title()}")
print(f"Membership    : {is_member}")
print(f"Monthly Spend : ₹{monthly_spend}")
print(f"Segment       : {segment}")
print(f"Type          : {customer_type}")
print(f"Priority City : {priority_city}")
```
