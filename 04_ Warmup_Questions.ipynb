```python
# 1) Reverse a string without using reverse()
 
language = "Python"

language[::-1]
```

```python
# 2) Count vowels in a string

role = "Data Science"

vowels = ["a", "e", "i", "o", "u"]

count = 0

for a in role:
    if a in vowels:
        count +=1
print(count)
```

```python
# 3) Check if a word is a palindrome


word = "madam"

word == word[::-1]
```

```python
# (ignore spaces + caps)

def is_pali(word):
    word = word.replace(" ", "").lower()
    return word == word[::-1]

print(is_pali("Race CAR"))
```

```python
# Check if two strings are anagrams

s1 = "listen"
s2 = "silent"

print(sorted(s1) == sorted(s2))
```

```python
# Count char frequency in string

text = "Banana"

freq = {}

for ch in text:
    freq[ch] = freq.get(ch, 0)+ 1

print(freq)
```
